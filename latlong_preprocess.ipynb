```python
import pandas as pd
import time
from os import listdir
from geopy.geocoders import Nominatim

geolocator = Nominatim(timeout = 20)

path_read = '/content/drive/Shareddrives/TCC 2022 - Marcelo/Atividade 1 - Pré processamento/5w1h - nyt dataset'
path_write = '/content/drive/Shareddrives/TCC 2022 - Marcelo/Atividade 1 - Pré processamento/5w1h - nyt dataset lat long'

pickles = listdir(path_read)
processed_pickles = set(listdir(path_write))

for pickle in pickles:
    if pickle in processed_pickles: continue

    data = pd.read_pickle(path_read + '/' + pickle)
    data['coords'] = None

    print(f'=============== Pickle: {pickle} ===============')

    notna_data = data[data['where'].notna()]
    
    start = time.time()

    for index, row in notna_data.iterrows():
        print(f'{index}/{len(data)} starting time: {time.time() - start}')
        
        address = geolocator.geocode(row['where'])
        if address is None: continue

        row['coords'] = (address.latitude, address.longitude)

        data.loc[index] = row
        
    end = time.time()

    data.to_pickle(path_write + '/' + pickle)
```
