##**Leitura dos pickles**

```python
path = '/content/drive/Shareddrives/TCC 2022 - Marcelo/Atividade 1 - Pré processamento/5w1h - nyt dataset lat long'

pickles_names = os.listdir(path)
print(pickles_names)
```

```python
pickles_dates = []
for pickle_name in pickles_names:
    pickle_name = pickle_name.replace('.', '_').split('_')
    pickles_dates.append((int(pickle_name[1]), int(pickle_name[2])))

print(pickles_dates)
```

```python
pickles = sorted(zip(pickles_dates, pickles_names), key=lambda x: x[0])
print(pickles)
```

```python
month_data = [pd.read_pickle(path + '/' + pickle_name) for pickle_date, pickle_name in pickles]      
```

```python
data = pd.concat(month_data).reset_index().drop('index', axis = 1)              #concatenando todos os pickles dos meses
data = data[data.notnull()['coords']].reset_index().drop('index', axis = 1)     #removendo eventos sem representacao geografica (sem coordenadas (lagitude, longitude))

data
```

```python
pre_removal = data.isna().sum()
```

##**Remoção de componentes stopwords**

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer

def is_stopword_component(component_text):
    tokenizer = RegexpTokenizer(r'\w+')

    stop_words = set(stopwords.words('english'))
    words = tokenizer.tokenize(component_text)

    for word in words:
        if word.lower() not in stop_words:
            return False

    return True
```

```python
for index, row in data.iterrows():
    for component in ['what', 'when', 'who', 'why', 'how']:
        if row[component] is not None and is_stopword_component(row[component]):
            row[component] = None
```

```python
post_removal = data.isna().sum()

missing_values_components = pd.DataFrame({'Pre Remocao': pre_removal, 'Pos Remocao': post_removal}).loc[['text','what', 'when', 'where', 'who', 'why', 'how']]
missing_values_components = missing_values_components.style.set_caption('Componentes Nulas')
missing_values_components
```
